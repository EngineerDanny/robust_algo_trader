```python
import pandas as pd
from glob import glob

date_time = "2024-01-03_05:09"
date_time = "2024-01-03_05:21"
date_time = "2024-01-03_05:28"
date_time = "2024-01-03_05:39"
# date_time = "2024-01-03_05:43"

date_time = "2024-01-03_07:42"




out_df_list = []
for out_csv in glob(f"/scratch/da2343/ml_project_2_{date_time}/results/*.csv"):
    out_df_list.append(pd.read_csv(out_csv))
return_df = pd.concat(out_df_list, axis=1)
# remove duplicate columns
return_df = return_df.loc[:, ~return_df.columns.duplicated()]

# fill in missing values with 0
return_df = return_df.fillna(0)

# root_results_dir = "/projects/genomic-ml/da2343/ml_project_2/cnn/results"
# return_df.to_csv(f"{root_results_dir}/{date_time}_results.csv", index=False)

# return_df['Mean Return'] = return_df.iloc[:, 1:].sum(axis=1)
# return_df = return_df[['Month', 'Mean Return']]
return_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = return_df


# Plot each currency pair's percent return over the years
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(df['Year'], df['Mean Return'])

# Customize the plot
ax.set_xlabel('Year')
ax.set_ylabel('Percent Return')
ax.set_title('Currency Pairs Percent Return Over the Years')
ax.legend()
plt.grid(True)
plt.show()
```
