```python
# Import the libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv("/projects/genomic-ml/da2343/ml_project_2/cnn/results/2023-12-23_13:34_results.csv")

# Group the data by the dataset_name column
groups = data.groupby("dataset_name")
```

```python


# Create a figure and a set of subplots for each group
for name, group in groups:
    fig, axes = plt.subplots(2, 2, figsize=(10, 10))
    fig.suptitle(f"Data for {name}") # Add a title for the figure

    # Plot the Return vs Year as a line plot
    axes[0, 0].plot(group["Year"], group["Return"])
    axes[0, 0].set_title("Return vs Year")
    axes[0, 0].set_xlabel("Year")
    axes[0, 0].set_ylabel("Return")

    # Plot the trades vs Year as a bar plot
    axes[0, 1].bar(group["Year"], group["trades"])
    axes[0, 1].set_title("trades vs Year")
    axes[0, 1].set_xlabel("Year")
    axes[0, 1].set_ylabel("trades")

    # Plot the Max Drawdown vs Year as a scatter plot
    axes[1, 0].scatter(group["Year"], group["Max Drawdown"])
    axes[1, 0].set_title("Max Drawdown vs Year")
    axes[1, 0].set_xlabel("Year")
    axes[1, 0].set_ylabel("Max Drawdown")

    # Plot the Percent Return vs Year as a histogram
    axes[1, 1].hist(group["Percent Return"])
    axes[1, 1].set_title("Percent Return vs Year")
    axes[1, 1].set_xlabel("Percent Return")
    axes[1, 1].set_ylabel("Frequency")

    # Adjust the spacing between the subplots
    plt.tight_layout()

    # Show the figure
    plt.show()
```
